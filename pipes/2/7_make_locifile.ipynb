```python
import pandas as pd
```

```python
path_tes = '/home/juan/Desktop/juan/bio/mrcv/data/TEs/blast/filtered/all.gff3'
path_genes = '/home/juan/Desktop/juan/bio/mrcv/data/genome/genes.gff3'
path_promoters = '/home/juan/Desktop/juan/bio/mrcv/data/genome/promoters.gff3'
path_locifile = '/home/juan/Desktop/juan/bio/mrcv/data/genome/GPT.locifile.csv'
path_locifile_demo = '/home/juan/Desktop/juan/bio/mrcv/data/genome/GPT.locifile.demo.csv'
```

```python
df_tes = pd.read_csv(path_tes, sep='\t')
df_tes.columns = ['seqname' , 'source' , 'feature' , 'start' , 'end' , 'score' , 'strand' , 'frame' , 'attribute']

print(len(df_tes.index))
df_tes.head(5)
```

```python
df_genes = pd.read_csv(path_genes, sep='\t')
df_genes.columns = ['seqname' , 'source' , 'feature' , 'start' , 'end' , 'score' , 'strand' , 'frame' , 'attribute']

print(len(df_genes.index))
df_genes.head(5)
```

```python
df_promoters = pd.read_csv(path_promoters, sep='\t')
df_promoters.columns = ['seqname' , 'source' , 'feature' , 'start' , 'end' , 'score' , 'strand' , 'frame' , 'attribute']

print(len(df_promoters.index))
df_promoters.head(5)
```

```python
df_res = pd.concat([df_tes, df_genes, df_promoters])
```

```python
print(len(df_res.index))
```

```python
df_res['locifile_position'] = df_res.seqname + ':' + df_res.start.astype(str) + '-' + df_res.end.astype(str)
```

```python
df_res.head(3)
```

```python
df_save = df_res[['locifile_position','attribute']]
```

```python
df_save.to_csv(path_locifile, header=None, index=None,sep='\t')
```

```python
df_save.sample(200000).to_csv(path_locifile_demo, header=None, index=None,sep='\t')
```

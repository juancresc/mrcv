```python
import pandas as pd
```

```python
path_target = '../../data/res/04_02_19/psRNATargetJob-1552408016123276.txt'
path_mirnas = '../../data/res/04_02_19/DEG.Y.csv'
path_deg_mir = '../../data/res/04_02_19/diffexpr-results.0.1.csv'
path_deg_genes  = '../../data/res/04_02_19/DEG-normalized.csv'
path_merged = '../../data/res/04_02_19/targets_go.csv'
path_targets_out = '../../data/res/04_02_19/targets.csv'
path_taes = '../../data/GO/taes.csv'
path_taes_2 = '../../data/GO/taes1.0.TAB'
path_aet = '../../data/GO/aet.csv'
path_ath = '../../data/GO/ath.csv'
```

```python
df_target = pd.read_csv(path_target, sep='\t', comment='#')
print(len(df_target.index))
df_target.rename(columns={'miRNA_Acc.':'miRNA_name','Target_Acc.':'target'}, inplace=True)
df_target = df_target[['miRNA_name','target','Expectation','miRNA_aligned_fragment','Target_aligned_fragment','Inhibition']]
df_target.head(3)
```

```python
#create target gene column in df_psrna
df_target['gene'] = df_target['target'].str.split('.').str[0]
df_target.head(2)
```

## add l2fc miRNA

```python
df_deg_mirna = pd.read_csv(path_deg_mir)
```

```python
df_deg_mirna = df_deg_mirna[['Gene','log2FoldChange']]
```

```python
df_deg_mirna.head(2)
```

```python
#merge df
df_merged_mirna = pd.merge(df_target, df_deg_mirna, left_on='miRNA_name', right_on='Gene', how='left')
print(len(df_merged_mirna.index))
```

```python
df_merged_mirna.rename(columns={'log2FoldChange':'miRNA_l2fc'}, inplace=True)
df_merged_mirna.drop('Gene', axis=1, inplace=True)
```

```python
df_merged_mirna.head(3)
```

## add l2fc gene

```python
df_deg_genes = pd.read_csv(path_deg_genes)
```

```python
df_deg_genes = df_deg_genes[['gene','log2FoldChange']]
df_deg_genes['gene'] = df_deg_genes['gene'].str.split(':').str[1]
```

```python
df_deg_genes.head(2)
```

```python
#merge df
df_merged_mirna_gene = pd.merge(df_merged_mirna, df_deg_genes, left_on='gene', right_on='gene', how='left')
print(len(df_merged_mirna_gene.index))
```

```python
df_merged_mirna_gene.rename(columns={'log2FoldChange':'gene_l2fc'}, inplace=True)
```

```python
df_merged_mirna_gene.head(3)
```

```python
cols = ['miRNA_name','miRNA_l2fc','target','gene','gene_l2fc','Expectation','miRNA_aligned_fragment','Target_aligned_fragment','Inhibition']
df_merged_mirna_gene = df_merged_mirna_gene[cols]
```

```python
df_merged_mirna_gene.head(2)
```

## add miRNA mature sequence

```python
df_mirnas = pd.read_csv(path_mirnas, sep='\t')
df_mirnas = df_mirnas[['Name','MajorRNA','#Locus']]
df_mirnas.head(2)
```

```python
df = pd.merge(df_merged_mirna_gene,df_mirnas, left_on='miRNA_name',right_on='Name',how='left' )
df.drop('Name', axis=1, inplace=True)
```

```python
df.head(3)
```

### triticum

```python
df_taes = pd.read_csv(path_taes_2,sep='\t')
df_taes = df_taes[['Gene-ID','Human-Readable-Description']]
df_taes.rename(columns={'Gene-ID':'gene','Human-Readable-Description':'description'}, inplace=True)
df_taes.head(5)
```

```python
df_taes[df_taes.gene == 'TraesCS6B01G248300.1']
```

```python
df_taes['gene'] = df_taes['gene'].str.replace('A01','A02')
df_taes['gene'] = df_taes['gene'].str.replace('B01','B02')
df_taes['gene'] = df_taes['gene'].str.replace('D01','D02')
```

```python
df_taes.head(2)
```

```python
df_taes[df_taes.gene == 'TraesCS6B02G248300.1']
```

```python
df_merged_mirna_gene_taes = pd.merge(df_merged_mirna_gene,df_taes,left_on='target',right_on='gene',how='left')
df_merged_mirna_gene_taes.rename(columns={'gene_x':'gene'}, inplace=True)
df_merged_mirna_gene_taes.drop('gene_y', axis=1, inplace=True)
df_merged_mirna_gene_taes.head()
```

```python
df_merged_mirna_gene_taes.to_csv(path_merged,index=None)
```

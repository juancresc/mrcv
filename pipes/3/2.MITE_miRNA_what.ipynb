```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_all = '../../data/ss/04_02_19/Results.txt'
path_mites = '../../data/ss/mites_hp/Results.csv'
```

```python
df_all = pd.read_csv(path_all, sep='\t')
print(len(df_all.index))
counts  = df_all.MIRNA.value_counts()
#city_count = city_count[:10,]
plt.figure(figsize=(10,5))
sns.barplot(counts.index, counts.values, alpha=0.8)
plt.title('A')
plt.ylabel('miRNAs', fontsize=12)
plt.xlabel('N', fontsize=12)
plt.show()
```

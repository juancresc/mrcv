```python
import pandas as pd
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
from subprocess import Popen, PIPE, call
path_target_finder = '/home/juan/Desktop/juan/bio/mrcv/sw/TargetFinder/targetfinder.pl'
target_seqs = '../../data/res/DEG-cdna.fasta'
file_deg_mirna_fasta = '../../data/res/DEG-miRNA.fasta'
```

```python
df = pd.read_csv('../../data/ss/04_02_19/DEG.Y.csv', sep='\t')
print(len(df.index))
df.head(3)
```

```python
records = []
for k,v in df.iterrows():
    record = SeqRecord(Seq(v.MajorRNA), id=v.Name, description=str(v.log2FoldChange))
    records.append(record)
SeqIO.write(records, file_deg_mirna_fasta, 'fasta')
```

```python
#search targets
out_file = open("../../data/res/04_02_19/targets_cdna.csv","w")
print(len(df.index))
out_total = ''
for k,v in df.iterrows():
    print(k,)
    if 'N' in v.MajorRNA:
        print(v.Name + "has Ns")
        continue
    cmd_list = ['perl', path_target_finder,'-s',v.MajorRNA,'-q',v.Name,'-d',target_seqs,'-p','table']
#    print(' '.join(cmd_list))
    pro = Popen(cmd_list, stdout=PIPE, stderr=PIPE)
    out,err = pro.communicate()
#    out = str(out)
#    err = str(err)
    print(out,err)
    if b"No results for" in out:
        continue
#    out_total += out.decode("utf-8") 
    out_file.write(out.decode("utf-8") )
out_file.close()
#os.system('grep -v "No results for" targets.csv > targets.clean.csv')
```

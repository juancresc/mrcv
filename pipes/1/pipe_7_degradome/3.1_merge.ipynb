```python
import pandas as pd
```

```python
path_cleaves = '../../../data/res/04_02_19/cleaveland/'
path_cleaves_res = '../../../data/res/04_02_19/cleaves.csv'
path_compare_res = '../../../data/res/04_02_19/compared_res.csv'
path_merged = '../../../data/res/04_02_19/compared_res_degradome.csv'
degs = ['SRR3690386','SRR6328740','SRR6328739','SRR1197126','SRR1197128','SRR1197127','SRR8113357','SRR1197125','SRR475422','SRR475423','SRR3710487']
```

```python
dfs = []
for deg in degs:
    df = pd.read_csv(path_cleaves + deg + '/cleaves.csv', sep='\t', comment='#')
    df['degradome'] = deg
    dfs.append(df)
df_all = pd.concat(dfs)
df_all['Query'] = df_all['Query'].str.split(' ').str[0]
```

```python
df_all = df_all[['Query','Transcript','DegradomeCategory','DegradomePval','degradome']]
print(len(df_all.index))
df_all.head(2)
```

```python
len(df_all[['Query','Transcript']].drop_duplicates().index)
```

```python
df_all.to_csv(path_cleaves_res, sep='\t', index=None)
df_all
```

```python
df_res = pd.read_csv(path_compare_res, sep='\t', comment='#')
```

```python

print(len(df_res.index))
df_res.head(2)
```

```python
df_merge = pd.merge(df_res, df_all, left_on=['miRNA_name','target'], right_on=['Query','Transcript'], how='left')
```

```python
df_merge.drop(columns=['Query', 'Transcript'], inplace=True)
```

```python
df_merge.head(2)
```

```python
df_merge.to_csv(path_merged, sep='\t', index=None)
path_merged
```

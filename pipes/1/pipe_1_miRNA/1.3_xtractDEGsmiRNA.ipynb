```python
import pandas as pd
path_data_dir = "../../data/ss/04_02_19/"
path_res_dir = "../../data/res/04_02_19/"
path_deg_file = path_res_dir + "diffexpr-results.0.1.csv"
```

```python
path_results = path_data_dir + "Results.txt"
df_res = pd.read_csv(path_results, sep="\t")
df_res.head(2)
```

```python
df_res = df_res[df_res.MIRNA == "Y"]
print(len(df_res.index))
```

```python
df_res.to_csv(path_data_dir + "results.valid.csv", sep="\t", index=None)
```

```python
df_deg = pd.read_csv(path_deg_file, sep=",")
df_deg.head(2)
```

```python
df_deg = df_deg[['Gene','log2FoldChange']]
```

```python
df_deg_y = pd.merge(df_res, df_deg, left_on='Name', right_on='Gene')
df_deg_y.head(2)
```

```python
df_deg_y.to_csv(path_res_dir + "DEG.Y.csv", sep="\t", index=None)
```

```python
import pandas as pd
```

```python
path_deg_tgac = "../../data/compare/deg_tgac.csv"
path_map = "../../data/compare/map_tgac.csv"
path_res = '../../data/res/04_02_19/targets_go_mirbase.csv'
path_compare_res = '../../data/res/04_02_19/compared_res.csv'
```

```python
df_deg_tgac = pd.read_csv(path_deg_tgac)
#df_deg_tgac = df_deg_tgac[['ensembl_gene_ID','log2FoldChange']]
df_deg_tgac.head(2)
```

```python
df_map = pd.read_csv(path_map)
```

```python
df_map['Matched ID(s)'] = df_map['Matched ID(s)'].str.replace('A01','A02')
df_map['Matched ID(s)'] = df_map['Matched ID(s)'].str.replace('B01','B02')
df_map['Matched ID(s)'] = df_map['Matched ID(s)'].str.replace('D01','D02')
```

```python
df_map.head(2)
```

```python
## get results file
```

```python
df_res = pd.read_csv(path_res, sep="\t")
df_res.head(2)
```

```python
#df_res = df_res[['gene','gene_l2fc']]
df_res.head(2)
```

```python
## merge result with map
```

```python
df_res_mapped = pd.merge(df_res,df_map,left_on='gene',right_on='Matched ID(s)', how='left')
```

```python
df_res_mapped.head(2)
```

```python
df_res_mapped.drop('Matched ID(s)',axis=1,inplace=True)
```

```python
## add l2fc tgac
```

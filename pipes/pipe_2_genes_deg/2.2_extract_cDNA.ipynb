```python
import pandas as pd
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
```

```python
#IWGSC
file_deg = '../../data/res/04_02_19/DEG-normalized.csv'
file_deg_cdna = '../../data/res/04_02_19/DEG-cdna.fasta'
file_cdna = "/home/juan/Desktop/juan/bio/data/IWGSC/42/Triticum_aestivum.IWGSC.cdna.all.fa"
gene_id = 'ID='
df_list = pd.read_csv(file_deg, index_col=False, sep=',', comment='#')
genes = df_list.gene.tolist()
genes = [gen.split(':')[1] for gen in genes]
print(len(genes))
```

```python
seqs = SeqIO.parse(file_cdna,'fasta')
records_out = []
for record in seqs:
    gene_name = record.id.split('.')[0]
    record.description = ""
    if gene_name in genes:
        records_out.append(record)
SeqIO.write(records_out, file_deg_cdna, 'fasta')
records_out
```

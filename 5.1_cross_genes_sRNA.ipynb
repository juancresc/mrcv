```python
import pandas as pd
```

```python
df_srna = pd.read_csv('data/res/sRNA_DEG_mirbase.csv', sep='\t')
df_srna.head(2)
```

```python
df_target = pd.read_csv('data/res/targets.clean.csv', sep='\t', header=None)
df_target.columns = ['target_srna','target_gene','target_from','target_to','target_strand','target_mismatches','target_srna_seq','target_aln','target_gene_seq']
df_target.head(2)
```

```python
df_res = pd.merge(df_srna, df_target, left_on='Name', right_on='target_srna', how='left')
df_res.head()
```

```python
df_res['target_gene'] = df_res['target_gene'].str.split(' ').str[0]
df_res.head()
```

```python
df_mart_traes = pd.read_csv('data/data/marts/mart_export_traes.txt', sep='\t')
df_mart_traes.head()
```

```python
df_mart_traes_at = pd.read_csv('data/data/marts/mart_export_traes_at.txt', sep='\t')
df_mart_traes_at.head()
```

```python
df_mart_at = pd.read_csv('data/data/marts/mart_export_at.txt', sep='\t')
df_mart_at.head()
```

```python
df_traes_link_at = pd.merge(df_mart_traes, df_mart_traes_at, on='Gene stable ID', how='left')
df_traes_link_at.head(4)
```

```python
df_traes_at = pd.merge(df_traes_link_at, df_mart_at, left_on='Arabidopsis thaliana gene stable ID', right_on='Gene stable ID', how='left')
df_traes_at.head()
```

```python
df_res_mart = pd.merge(df_res, df_traes_at, left_on='target_gene', right_on='Gene stable ID_x', how='left')
df_res_mart.head()
```

```python
df_res_mart.to_csv('data/res/srna_targets.csv', sep='\t', index=None)
```

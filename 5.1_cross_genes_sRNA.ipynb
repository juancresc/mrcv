```python
import pandas as pd
import numpy as np
```

```python
df_srna = pd.read_csv('data/res/sRNA_DEG_mirbase.csv', sep='\t')
df_srna.head(2)
```

```python
df_target = pd.read_csv('data/res/targets_dna.clean.csv', sep='\t', header=None)
df_target.columns = ['target_srna','target_gene','target_from','target_to','target_strand','target_mismatches','target_srna_seq','target_aln','target_gene_seq']
print(len(df_target.index))
df_target.head(2)
```

```python
df_target_genes = pd.merge(df_srna, df_target, left_on='Gene', right_on='target_srna', how='left')
print(len(df_target_genes.index))
df_target_genes.head()
```

```python
df_target_genes['target_gene'] = df_target_genes['target_gene'].str.split(' ').str[0]
df_target_genes.head()
```

```python
df_mart_traes = pd.read_csv('data/data/marts/mart_export_traes.txt', sep='\t')
df_mart_traes = df_mart_traes[['Gene stable ID', 'GO term name', 'GO term definition', 'GO domain','Gene description']]
df_mart_traes.columns = ['gene','GO','GO_description','GO_domain','gene_description']
df_mart_traes.head()
```

```python
df_mart_traes_at = pd.read_csv('data/data/marts/mart_export_traes_at.txt', sep='\t')
df_mart_traes_at.columns = ['gene','at_gene','at_gene_name']
df_mart_traes_at.head()
```

```python
df_mart_at = pd.read_csv('data/data/marts/mart_export_at.txt', sep='\t')
df_mart_at = df_mart_at[['Gene stable ID', 'GO term name', 'GO term definition', 'GO domain','Gene description']]
df_mart_at.columns = ['at_gene','at_GO','at_GO_definition','at_GO_domain', 'at_gene_description']
df_mart_at.head()
```

```python
df_res = pd.merge(df_mart_traes, df_mart_traes_at, left_on='gene', right_on='gene', how='left')
df_res = df_res.replace(np.nan, '', regex=True)

rules = {'GO':','.join,'GO_description':','.join, 'GO_domain':','.join, 'gene_description':','.join, 'at_gene':','.join, 'at_gene_name':','.join}
df_res = df_res.groupby('gene',as_index=False).agg(rules)
print(len(df_res.index))
df_res.head()
```

```python
df_res_2 = pd.merge(df_res, df_mart_at, left_on='at_gene', right_on='at_gene', how='left')
df_res_2 = df_res_2.replace(np.nan, '', regex=True)

rules = {'GO':','.join,'GO_description':','.join, 'GO_domain':','.join, 'gene_description':','.join, 'at_gene':','.join, 'at_gene_name':','.join, 'at_GO':','.join,'at_GO_definition':','.join,'at_GO_domain':','.join,'at_gene_description':','.join}
df_res_2 = df_res_2.groupby('gene',as_index=False).agg(rules)
print(len(df_res_2.index))
df_res_2.head()
```

```python
df_res_3 = pd.merge(df_target_genes, df_res_2, left_on='target_gene', right_on='gene', how='left')
df_res_3 = df_res_3.replace(np.nan, '', regex=True)
df_res_3.head(5)
```

```python
df_res_3.columns
```

```python
rules = {}
rules['MajorRNA'] = 'first'
rules['log2FoldChange'] = 'first'
rules['miRBASE'] = 'first'
rules['gene'] = ','.join
rules['gene_description'] = ','.join
rules['GO_description'] = ','.join
rules['GO_domain'] = ','.join
rules['at_GO'] = ','.join
rules['at_gene_name'] = ','.join
rules['at_GO_definition'] = ','.join
rules['at_gene_description'] = ','.join
rules['at_GO_domain'] = ','.join
rules['at_gene'] = ','.join

#rules['GO'] = ['Gene','target_gene']
#rules = {'GO':','.join,'GO_description':','.join, 'GO_domain':','.join, 'gene_description':','.join, 'at_gene':','.join, 'at_gene_name':','.join, 'at_GO':','.join,'at_GO_definition':','.join,'at_GO_domain':','.join,'at_gene_description':','.join}
df_res_4 = df_res_3.groupby(['Gene','target_gene'],as_index=False).agg(rules)
print(len(df_res_4.index))
df_res_4.head()
```

```python
df_res_4.to_csv('data/res/targets_mart.csv', sep='\t', index=None)
```
